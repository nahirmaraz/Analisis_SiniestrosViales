### Análisis multivariado.

Se realizan 3 gráficos, de frecuencia de roles, frecuencia de victimas y distribucion de edades, todos discrminados por sexo.

```python
# Crear un lienzo (figura) y organizar subgráficos en una cuadrícula de 1x2
fig, axs = plt.subplots(1, 2, figsize=(15, 6), sharey=True)
# Gráfico de barras apiladas para roles distinguidos por sexo
roles_por_sexo = df_vict.groupby(['ROL','SEXO']).size().unstack()
roles_por_sexo_por_sexo_por_sexo = roles_por_sexo[roles_por_sexo.sum().sort_values(ascending=False).index]
roles_por_sexo.plot(kind='bar', stacked=True, ax=axs[0], color=['orange', 'blue'])
axs[0].set_title('Roles por Sexo')
axs[0].set_ylabel('Cantidad')
# Gráfico de barras apiladas para víctimas distinguidas por sexo
victimas_por_sexo = df_vict.groupby(['VICTIMA','SEXO']).size().unstack()
victimas_por_sexo_por_sexo = victimas_por_sexo[victimas_por_sexo.sum().sort_values(ascending=False).index]
victimas_por_sexo.plot(kind='bar', stacked=True, ax=axs[1],color=['orange', 'blue'])
axs[1].set_title('Víctimas por Sexo')
axs[1].set_ylabel('Cantidad')
# Ajustar diseño para mostrar ambos gráficos
plt.tight_layout()
# Crear un nuevo lienzo para los histogramas de Seaborn
plt.figure(figsize=(12, 6))
# Gráfico de distribución para edades distinguidas por sexo
sns.histplot(data=df_vict[df_vict['EDAD'].apply(lambda x: x.isnumeric() if isinstance(x, str) else True)], x='EDAD', hue='SEXO', multiple='stack', bins=20, kde=True)
plt.title('Distribución de Edades por Sexo')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
# Mostrar el lienzo con los gráficos
plt.show()
```

Es evidente que en todos los casos los hombres son los maypores responsables y perjudicados a la vez. También se observa que la edad recurrente es entre 20-40. Los roles que mas impactan son los conductores y las mayores víctimas son las motocicletas, seguido del peatón.
